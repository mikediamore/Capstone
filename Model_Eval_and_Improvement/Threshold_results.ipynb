### Main Takeaways

1.) Used more relaxed defintions of a market event, ranging from 1.5 Standard Deviations to 3.5 in increments of .5

2.) Using a relaxed definiton we see a decrease in accuracy, because there is more of a balance in the classes. We see an increase in F1 Scores as it is easier to predict positive class. What's most revealing are the plots themselves and the list of indices

    a.) As we can see from the list of intersections, the plot of predicted events and the plot of actual events that no matter what restriction we place on the market event definition we still have similar behavior. It will predict before and after a market event occurs. In this sense we always overpredict the number of events
    
    b.) However, even in relaxed defintions of the market event it seems like the predictions are clumped around certain areas in the time series
